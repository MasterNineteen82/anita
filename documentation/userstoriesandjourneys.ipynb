## User Stories and User Journeys

This section provides user stories and user journeys to further illustrate the interactions with the ANITA system. Visualizations will be added to enhance understanding.

### 1. Card Management

**User Story 1: As an End User, I want to read data from a card so that I can verify its contents.**

*   **User Journey:**
    1.  The End User approaches the card reader with a card.
    2.  The system detects the card (CM-01).
    3.  The End User selects the "Read Card Data" option in the UI.
    4.  The system reads the data from the card (CM-02).
    5.  The system displays the card data in a formatted manner.
    6.  The End User verifies the card contents.

**User Story 2: As an Administrator, I want to block a card so that it cannot be used for future access.**

*   **User Journey:**
    1.  The Administrator logs into the system.
    2.  The Administrator navigates to the Card Management section.
    3.  The Administrator selects the "Block Card" option.
    4.  The Administrator enters the card ID to be blocked.
    5.  The system blocks the card (CM-06).
    6.  The system confirms that the card has been blocked.

### 2. UWB Positioning

**User Story: As an End User, I want to view the location of a UWB device so that I can track its whereabouts.**

*   **User Journey:**
    1.  The End User logs into the system.
    2.  The End User navigates to the UWB Positioning section.
    3.  The system displays the current location of registered UWB devices (UWB-02).
    4.  The End User views the location of the desired device on a map or list.

**User Story: As an Administrator, I want to register a new UWB device so that it can be tracked by the system.**

*   **User Journey:**
    1.  The Administrator logs into the system.
    2.  The Administrator navigates to the UWB Management section.
    3.  The Administrator selects the "Register Device" option.
    4.  The Administrator enters the device details (ID, name, etc.).
    5.  The system registers the new UWB device (UWB-01).
    6.  The system confirms that the device has been registered.

### 3. Biometrics

**User Story: As an End User, I want to authenticate using my fingerprint so that I can gain secure access.**

*   **User Journey:**
    1.  The End User approaches the biometric device.
    2.  The system prompts for fingerprint authentication.
    3.  The End User places their finger on the fingerprint scanner.
    4.  The system authenticates the user via fingerprint (BIO-03).
    5.  If authentication is successful, the system grants access.

**User Story: As an Administrator, I want to manage the biometric database so that I can maintain accurate user records.**

*   **User Journey:**
    1.  The Administrator logs into the system.
    2.  The Administrator navigates to the Biometrics Management section.
    3.  The Administrator selects the "Manage Biometric Database" option.
    4.  The Administrator can view, add, or remove user biometric data (BIO-05).
